# Netflix Data Analysis Project

## Problem Statement
The objective of this project is to analyze Netflix’s movies and TV shows dataset to identify trends in content type, release patterns, country-wise distribution, ratings, and duration using data cleaning and exploratory data analysis techniques.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Fill missing categorical values
for col in ['director', 'cast', 'country', 'rating']:
    df[col] = df[col].fillna('Unknown')

# Clean duration column
df['duration_value'] = df['duration'].str.extract('(\d+)').astype(float)
df['duration_type'] = df['duration'].str.extract('(min|Season|Seasons)')
```

```python
df
```

```python
df.to_csv("netflix_cleaned.csv", index=False)
```

## Exploratory Data Analysis (EDA) 
This section explores trends and patterns in Netflix content using visualizations and statistical analysis. where

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(
    kind='bar',                      # bar chart
    title='Distribution of Movies vs TV Shows',
    xlabel='Content Type',            # x-axis label
    ylabel='Count'                    # y-axis label
)
plt.legend()

# Show the plot
plt.show()
```

```python
# This cell extracts only the year from 'date_added'
# We do this to analyze year-wise trends

df['year_added'] = df['date_added'].dt.year

# Check how many titles were added each year
df['year_added'].value_counts().sort_index()
```

```python
# This cell visualizes how Netflix content grew over time

df['year_added'].value_counts().sort_index().plot(
    kind='line',                     # line chart for trend analysis
    title='Netflix Content Added Over Years',
    xlabel='Year',
    ylabel='Number of Titles'
)

plt.show()
```

```python
# This cell finds the top 10 countries producing Netflix content

top_countries = df['country'].value_counts().head(10)

# Display the result
top_countries
```

```python
# This cell visualizes top 10 countries using a bar chart

top_countries.plot(
    kind='bar',
    title='Top 10 Content Producing Countries on Netflix',
    xlabel='Country',
    ylabel='Number of Titles'
)

plt.show()
```

```python
# This cell checks how content is distributed across age ratings

df['rating'].value_counts()
```

```python
# This cell visualizes rating distribution

df['rating'].value_counts().plot(
    kind='bar',
    title='Netflix Content Rating Distribution',
    xlabel='Rating',
    ylabel='Count'
)

plt.show()
```

```python
# Filter only movies from the dataset
movies = df[df['type'] == 'Movie']

# Describe gives summary statistics like mean, min, max
movies['duration_value'].describe()
```

```python
# This cell shows distribution of movie durations

movies['duration_value'].plot(
    kind='hist',
    bins=30,                         # number of bins in histogram
    title='Distribution of Movie Durations',
    xlabel='Duration (Minutes)'
)

plt.show()
```

## Key Insights

- Netflix has more Movies than TV Shows, but TV Shows form a strong part of the catalog.
- Content additions increased significantly after 2015.
- The United States, India, and the United Kingdom dominate content production.
- Most Netflix content is rated TV-MA, indicating a focus on mature audiences.
- The majority of movies have runtimes between 80 and 120 minutes.
